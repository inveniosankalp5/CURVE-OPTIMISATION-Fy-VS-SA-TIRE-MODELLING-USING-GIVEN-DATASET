```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import scipy
```

```python
df=pd.read_excel('Tire_Modeling_Ex1.xlsx')
```

```python
x=df.SA
y=df.Fy
```

```python
plt.scatter(x,y)
```

```python
plt.show()
```

```python
from scipy import optimize

def test_func(x,d,c,b,e):
    return d * np.sin(c * np.arctan(b*x-e*(b*x)))

params, params_covariance = optimize.curve_fit(test_func, x, y,maxfev=1000)
                                               

print(params)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Data')
plt.plot(x, test_func(x, params[0], params[1],params[2],params[3]),'r-+',label='Fitted function')
plt.xlim(-25,25)
plt.legend(loc='best')
plt.show()
```
